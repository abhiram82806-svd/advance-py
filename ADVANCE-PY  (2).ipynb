```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if len(s1) != len(s2):
    print("Not anagrams")
else:
    flag = True
    for ch in set(s1):
        c1 = 0
        c2 = 0

        for x in s1:
            if x == ch:
                c1 = c1 + 1
        for y in s2:
            if y == ch:
                c2 = c2 + 1

        if c1 != c2:
            flag = False
            break

    if flag:
        print("Anagrams")
    else:
        print("Not anagrams")
```

STUDENT ENROLLMENT SYSTEM WITH ATTENDANCE SYSTEM

Course limit

If attendance <75 notify


```python
students = {}
courses = {}
enrollments = {}

# ADD STUDENT
def add_student():
    sid = input("Enter Student ID: ")
    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    sec = input("Enter Section: ")

    students[sid] = {
        "rollno": roll,
        "name": name,
        "section": sec
    }
    print("Student added successfully")

# ADD COURSE
def add_course():
    cid = input("Enter Course ID: ")
    cname = input("Enter Course Name: ")
    limit = int(input("Enter Course Limit: "))

    courses[cid] = {
        "course_name": cname,
        "max_students": limit,
        "enrolled": []
    }
    print("Course added successfully")

# ENROLL STUDENT
def enroll_student():
    sid = input("Enter Student ID: ")
    cid = input("Enter Course ID: ")

    if sid not in students or cid not in courses:
        print("Invalid Student or Course")
        return

    if len(courses[cid]["enrolled"]) >= courses[cid]["max_students"]:
        print("Course limit reached")
        return

    courses[cid]["enrolled"].append(sid)
    enrollments[(sid, cid)] = {
        "attendance": 0,
        "grade": None
    }
    print("Student enrolled successfully")

# MARK ATTENDANCE
def mark_attendance():
    sid = input("Student ID: ")
    cid = input("Course ID: ")
    att = int(input("Enter Attendance %: "))

    enrollments[(sid, cid)]["attendance"] = att

    if att < 75:
        print("⚠ Defaulter: Attendance below 75%")

# ASSIGN GRADE
def assign_grade():
    sid = input("Student ID: ")
    cid = input("Course ID: ")
    grade = input("Enter Grade: ")

    enrollments[(sid, cid)]["grade"] = grade
    print("Grade assigned")

# VIEW STUDENT REPORT
def student_report():
    sid = input("Enter Student ID: ")

    print("\nStudent Report")
    print(students[sid])

    for (s, c), data in enrollments.items():
        if s == sid:
            print("Course:", courses[c]["course_name"],
                  "| Attendance:", data["attendance"],
                  "| Grade:", data["grade"])

# COURSE WISE REPORT
def course_report():
    cid = input("Enter Course ID: ")

    print("\nCourse:", courses[cid]["course_name"])
    for sid in courses[cid]["enrolled"]:
        data = enrollments[(sid, cid)]
        print(students[sid]["name"],
              "| Attendance:", data["attendance"],
              "| Grade:", data["grade"])

# MAIN MENU
while True:
    print("""
1. Add Student
2. Add Course
3. Enroll Student
4. Mark Attendance
5. Assign Grade
6. View Student Report
7. Course Wise Report
8. Exit
""")

    ch = input("Enter choice: ")

    if ch == "1":
        add_student()
    elif ch == "2":
        add_course()
    elif ch == "3":
        enroll_student()
    elif ch == "4":
        mark_attendance()
    elif ch == "5":
        assign_grade()
    elif ch == "6":
        student_report()
    elif ch == "7":
        course_report()
    elif ch == "8":
        print("Exiting system")
        break
    else:
        print("Invalid choice")
```

```python
import re

password = input("Enter your password: ")

strength = 0


if len(password) >= 8:
    strength += 1
if re.search("[A-Z]", password):
    strength += 1
if re.search("[a-z]", password):
    strength += 1
if re.search("[0-9]", password):
    strength += 1
if re.search("[@#$%^&*()!]", password):
    strength += 1
if strength == 5:
    print("Strong Password")
elif strength >= 3:
    print("Medium Password")
else:
    print("Weak Password")
```

```python
students = {}

def add_student():
    roll = input("Enter Roll Number: ")
    name = input("Enter Name: ")
    marks = []

    for i in range(3):
        m = int(input(f"Enter mark {i+1}: "))
        marks.append(m)

    students[roll] = {
        "name": name,
        "marks": marks
    }

    print("Student Added Successfully")


def calculate_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 40:
        return "D"
    else:
        return "Fail"


def show_report():
    roll = input("Enter Roll Number: ")

    if roll in students:
        student = students[roll]
        total = sum(student["marks"])
        avg = total / len(student["marks"])
        grade = calculate_grade(avg)

        print("\nReport Card")
        print("Name:", student["name"])
        print("Marks:", student["marks"])
        print("Total:", total)
        print("Average:", avg)
        print("Grade:", grade)
    else:
        print("Student not found")


def menu():
    while True:
        print("\n1. Add Student")
        print("2. Show Report")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            show_report()
        elif choice == "3":
            print("Goodbye")
            break
        else:
            print("Invalid choice")


menu()
```
